```python
def get_sub_tree(tree_df, oid):
    itself = tree_df[tree_df.child_oid==oid]
    subtree = set([(oid, itself.child_otype.values[0])])
    children_oid = tree_df[tree_df.parent_oid==oid].child_oid.values
    children_otype = tree_df[tree_df.parent_oid==oid].child_otype.values
    subtree.update(zip(children_oid, children_otype))
    if len(children_oid)==0:
        return subtree, itself.shape

    for coid in children_oid:
        coid_subtree, coid_shape = get_sub_tree(tree_df, coid)
        subtree.update(coid_subtree)
    return subtree, itself.shape
    

get_sub_tree(CHILD_PARENT, 41964978)
```

```python
sub_trees = []
for idx, row in CHILD_PARENT[CHILD_PARENT.parent=='root'].iterrows():
    oid = row.child_oid
    subtree, shape = get_sub_tree(CHILD_PARENT,oid)
    if shape!=(1,7):
        print oid
        print subtree
        print shape
    sub_trees.append(subtree)
```

```python
with open('../data/osm_sub_trees.txt', 'w') as f:
    for subtree in sub_trees:
        f.write(str(subtree)+'\n')
```

```python
OSM_DC_BBOX_PATH = '../data/dc_bbox.osm'
OSM_DC_BBOX = osm_container(OSM_DC_BBOX_PATH)
print OSM_DC_BBOX.data_size()
```

```python
otype_str_to_class = {'Node': Node, 'Way':Way, 'Relation':Relation}
from collections import defaultdict
all_tags = []
for subtree in sub_trees:
    tags = defaultdict(list)
    for oid, otype in subtree:
        obj = OSM_DC_BBOX.get_osm_obj_by_id(otype_str_to_class[otype], oid)
        for tag, value in obj.tags.items():
            tags[tag].append(value)
    all_tags.append(tags)
```
