```python
covered_segs_idx = pd.unique(df_segs_clip.index_seg)
cnt_segs = len(covered_segs_idx)
cnt_segs
```

```python
bk_facs = pd.read_csv('../open_data_analysis/data/feature_bk_facs_dc.csv',index_col=0)
bk_facs_covered = bk_facs.loc[covered_segs_idx]
for col in bk_facs_covered.columns[1:]:
    no_feature = (bk_facs_covered[col].isnull()).sum()*100.0/cnt_segs
    print "{:<10} null: {}%".format(col, no_feature)
    d = (bk_facs_covered[col].value_counts()/cnt_segs*100.0).to_dict()
    for k, v in d.items():
        print '{}: {}%'.format(k,v)
    print
```

```python
poi = pd.read_csv('../open_data_analysis/data/feature_poi_dc.csv',index_col=0)
poi_covered = poi.loc[covered_segs_idx]
for col in poi.columns[1:]:
    no_feature = (poi_covered[col]==0).sum()*100.0/cnt_segs
    print "{:<25} null: {}%, has feature: {}%".format(col, no_feature, 100-no_feature)
```
