```python
import geopandas as gp
path_intxn_ph = 'data/segments_intxn_ph.geojson'
path_segs_ph = 'data/segments_ph.geojson'
```

```python
segs_ph = gp.read_file(path_segs_ph)
```

```python
intxn_ph = gp.read_file(path_intxn_ph)
```

```python
unique_node_intxn = []
for nid, grp in intxn_ph.groupby('NODE_ID'):
    geometry = grp.geometry.values[0]
    unique_node_intxn.append((nid,geometry.buffer(0.0001)))
unique_node_intxn = gp.GeoDataFrame(unique_node_intxn, columns=['NODE_ID','geometry'])
```
