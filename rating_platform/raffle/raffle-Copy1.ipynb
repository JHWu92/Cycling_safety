```python
import numpy as np

import pandas as pd

import datetime
import pickle
```

```python
ratings = pd.read_csv('../DB-backup/2017-11-29/Rating.csv', index_col=0)
users = pd.read_csv('../DB-backup/2017-11-29/Users.csv', index_col=0)
```

```python
users.email.fillna('', inplace=True)
nontest_users = users[(~users.email.str.contains('test'))&(~users.email.str.contains('guest'))]
nontest_uid = nontest_users.user_id
print(users.shape[0], nontest_users.shape[0])
```

```python
legit_ratings = ratings[ratings.uid.isin(nontest_uid)]
legit_ratings = legit_ratings.merge(users, left_on='uid', right_on='user_id')[['rid', 'uid', 'email_y']]
print(ratings.shape[0], legit_ratings.shape[0])
```

```python
print(datetime.datetime.now())
try:
    with open('state.pkl', 'rb') as f:
        state = pickle.load(f)
    print('state exists')
except:
    print('state not exists')    
    state = np.random.get_state()
    with open('state.pkl', 'wb') as f:
        pickle.dump(state, f)
    pass


sample_size = 12
while True:
    np.random.set_state(state)
    winners = legit_ratings.sample(sample_size)
    if winners.uid.nunique()!=12:
        sample_size+=1
        print(winners.uid.nunique(), 'sample_size++')
    else:
        break
print(winners.uid.nunique())
winners
```
